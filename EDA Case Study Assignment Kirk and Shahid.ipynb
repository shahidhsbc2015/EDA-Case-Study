## Binning Continuous Variables

```python
# Binning the AMT_INCOME_TOTAL based on quantiles

def income_category(x):
    if x <= 112500:
        return('Low Income')
    elif x > 112500 and x <= 147150:
        return('Average Income')
    elif x > 147150 and x <= 202500:
        return('Moderate Income')
    else:
        return('High Income')
ca['INCOME_CATEGORY'] = ca['AMT_INCOME_TOTAL'].apply(income_category)
```

```python
ca['AMT_INCOME_TOTAL'].describe()
```

```python
# Once the new column has been created, AMT_INCOME_TOTAL can be dropped

ca[['INCOME_CATEGORY','AMT_INCOME_TOTAL']].head(10)
```

```python
# Binning credit category by AMT_CREDIT based on quantiles

def credit_category(x):
    if x <= 270000:
        return('Low-End Credit')
    elif x > 270000 and x <= 513531:
        return('Moderate Credit')
    elif x > 513531 and x <= 808650:
        return('Moderately-High Credit')
    else:
        return('High-End Credit')
ca['CREDIT_CATEGORY'] = ca['AMT_CREDIT'].apply(credit_category)
```

```python
ca['AMT_CREDIT'].describe()
```

```python
ca['AMT_CREDIT'].describe()
```

```python
# Once the new column has been created, AMT_CREDIT can be dropped

ca[['CREDIT_CATEGORY','AMT_CREDIT']].tail(10)
```

Binning these two continuous variables to an interval with fewer representatives helps smoothen the data.
